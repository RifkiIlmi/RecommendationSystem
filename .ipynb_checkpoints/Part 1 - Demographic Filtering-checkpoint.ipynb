```python
import pandas as pd
```

```python
df = pd.read_csv("data/demographic.csv")
df
```

```python
# Sort 
df.sort_values("vote_average", ascending=False)
```

```python
# step 1 filter (duration, genre, runtime, etc)
genre = ["Animation"]
duration = (60, 150)
year = (2000, 2019)
topk = 20
```

```python
df = df[df.release_year.between(year[0], year[1]) & 
        df.runtime.between(duration[0], duration[1]) & 
        df[genre].all(axis=1)]

df.head()
```

```python
recommendation = df.sort_values("vote_average", ascending=False).head(topk)
recommendation
```

```python
# IMDB Weighted Rating 
# WR = Rv + Cm / v + m
# v = jumlah voting film tersebut
# m = syarat minimum vote (personalized/ quantile)
# R = rata-rata rating film itu
# C = rata-rata rating semua film
```

```python
def imdb_score(df, q=0.9):
    df = df.copy()
    
    m = df.vote_count.quantile(q)
    C = (df.vote_average * df.vote_count).sum() / df.vote_count.sum()
    
    df = df[df.vote_count >= m]
    df["score"] = df.apply(lambda x: (x.vote_average * x.vote_count + C*m) / (x.vote_count + m), axis=1)
    
    return df
```

```python
df = imdb_score(df)
```

```python
recommendation = df.sort_values("vote_average", ascending=False).head(topk)
recommendation
```
